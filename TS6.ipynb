### Ejercicio a:

$$ y(n)=x(n−3)+x(n−2)+x(n−1)+x(n)$$

$$ T(z) = z^{-3} + z^{-2} + z^{-1} + z^{-0}$$

```python
b = np.array([1, 1, 1, 1])  
a = np.array([1])           

w = np.linspace(0, 2 * np.pi, 1024)
z = np.exp(1j * w)

P = np.polyval(b[::-1], z**-1)
Q = np.polyval(a[::-1], z**-1)
H = P / Q

plt.figure(figsize=(12, 5))


xticks = np.arange(0, 2 * np.pi + 0.1, np.pi / 2)
xtick_labels = [r'$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']

plt.subplot(1, 2, 1)
plt.plot(w, np.abs(H), color='lime')
plt.title(r'$|H(e^{j\omega})|$')
plt.xlabel(r'$\omega$')
plt.ylabel(r'$|H(e^{j\omega})|$')
plt.xticks(xticks, xtick_labels)
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(w, np.angle(H), color='lime')
plt.title(r'$\angle H(e^{j\omega})$')
plt.xlabel(r'$\omega$')
plt.ylabel(r'Fase [rad]')
plt.xticks(xticks, xtick_labels)
plt.grid(True)

plt.tight_layout()
plt.show()
```

Observaciones: Hay un cero en $ \frac{\pi}{2} $ y otro en $\pi$ siendo $\pi$ Nyquist. Por eso el modulo de la tranferencia en esos valores es cero.
