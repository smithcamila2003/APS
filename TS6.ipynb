# Ejercicio a:

$$ y(n)=x(n−3)+x(n−2)+x(n−1)+x(n)$$

$$ T(z) = z^{-3} + z^{-2} + z^{-1} + z^{-0}$$

![Ejercicio A](EjercicioA.jpg)


```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt


b = [1, 1, 1, 1]      # Coeficientes del numerador del filtro
a = [1]               # Coeficientes del denominador del filtro

w, h = signal.freqz(b, a, worN=1000)


plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(w, np.abs(h))
plt.title("Módulo")
plt.xticks(np.arange(0, np.pi + 0.01, np.pi / 4), 
           [r'$0$', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$', r'$\frac{3\pi}{4}$', r'$\pi$'])
plt.xlabel("ω [rad/muestra]")
plt.ylabel("|T(e^{jω})|")
plt.grid()


plt.subplot(1, 2, 2)
plt.plot(w, np.angle(h))
plt.title("Fase")
plt.xticks(np.arange(0, np.pi + 0.01, np.pi / 4), 
           [r'$0$', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$', r'$\frac{3\pi}{4}$', r'$\pi$'])
plt.xlabel("ω [rad/muestra]")
plt.ylabel("∠T(e^{jω}) [rad]")
plt.grid()

plt.tight_layout()
plt.show()
```
